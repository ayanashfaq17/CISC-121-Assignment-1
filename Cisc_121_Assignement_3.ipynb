```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        sl = []
        tl = []

        for i in s:
            sl.append(i)

        for i in t:
            tl.append(i)

        count = 0
        while count < len(sl):
            if sl[0] == tl[0]:
                sl.pop(0)
                tl.pop(0)
                count = 0
            else:
                a = sl.pop(0)
                sl.append(a)
                count += 1

        if len(sl) == 0:
          return True
        else:
          return False
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        sl = []
        tl = []

        for i in s:
            sl.append(i)

        for i in t:
            tl.append(i)

        count = 0
        while count < len(sl):
            if sl[0] == tl[0]:
                sl.pop(0)
                tl.pop(0)
                count = 0
            else:
                a = sl.pop(0)
                sl.append(a)
                count += 1

        return len(tl) == 0
```

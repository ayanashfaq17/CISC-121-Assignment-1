```python
class Solution:
    def reverse(self, x: int) -> int:
        sx = str(x)
        rvs = []
        for i in range(len(sx)-1, 0, -1):
            rvs.append(sx[i])

        revint = ""
        for i in range(len(rvs)):
            revint.append("".join(rvs[i]))

        if revint < -2^31 or revint > (2^31)-1:
            return 0
        return 1
```

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        ind = []
        for i in range(len(nums)):
            for j in range(len(nums)):
                if nums[i] + nums[j] == target:
                    ind.append(i)
                    ind.append(j)
                    return ind
```

**Question 5 Part 1**
Given an string s, return true if its a palindrome, else return false\
Input: string s, Output: Boolean(true or false), Constraints: 1 <= s.length <= 2*10**5, s consists of only printable ASCII characters\
ex. Input: s = "A man, a plan, a canal: Panama", Output: true, Explanation: "amanaplanacanalpanama" is a palindrome.

Pseudo code:
sl = s.lower()\
slist =[]
for loop looping through string and appending to slist
loop through list and remove delemiters like "," ":" " "

join list and reverse using [::-1]\
compare reversed string with original using if statemt to return true or false



**Part 2** First solution

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        ls = s.lower()
        slist = []

        for ch in ls:
            slist.append(ch)

        for i in range(len(slist)):
            if slist[i] == "," or ":" or " ":
                slist.remove(slist[i])

        sword = "".join(slist)
        rword = sword[::-1]

        if sword == rword:
            return True
        return False
```

After looking at a leetcode solution, https://leetcode.com/problems/valid-palindrome/solutions/7197820/clean-simple-efficient-beginner-friendly-easy-explanation, I figured out that looping through list of s was going to bbe tricky and there is a function that can take away nonalphanumeric characters which I found from leetcode solution

**Part 3** Working solution

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        ls = s.lower()
        slist = []

        for i in ls:
            if i.isalnum():
                slist.append(i.lower())

        sclean = "".join(slist)
        rsclean = sclean[::-1]

        if sclean == rsclean:
            return True
        return False
```
